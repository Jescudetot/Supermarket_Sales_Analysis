```python
import pandas as pd
import matplotlib.pyplot as plt  # Importar matplotlib.pyplot
# Calcular ingresos totales por línea de productos
income_by_product_region = region_data.groupby('Product_Line')['Gross_Income'].sum()
# Crear la gráfica
income_by_product_region.plot(
    kind='bar', 
    figsize=(10, 6), 
    title='Ingresos Totales por Línea de Productos en Yangon', 
    ylabel='Ingresos Totales ($)'
)
plt.xticks(rotation=45, ha='right')  # Ajustar etiquetas
plt.savefig("Ingresos_Yangon.png")  # Guardar gráfica como imagen
plt.show()  # Mostrar la gráfica
```

```python
# Gráfica 3: Métodos de Pago
payment_methods_region = region_data['Payment'].value_counts()
payment_methods_region.plot(
    kind='pie', 
    figsize=(10, 6), 
    autopct='%1.1f%%', 
    title='Distribución de Métodos de Pago en Yangon'
)
plt.savefig("Metodos_Pago_Yangon.png")  # Guarda la gráfica
plt.show()  # Mostrar la gráfica
```
